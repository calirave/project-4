```python
# Import the modules
import pandas as pd
from pathlib import Path
import hvplot.pandas
from sklearn.cluster import KMeans
```

```python
# Read in the CSV file as a Pandas DataFrame
holiday_songs_spotify_df = pd.read_csv(
    Path("data/Holiday_Songs_Spotify.csv")
)

#C:\Users\wsabr\OneDrive\Desktop\project-4\data\Holiday_Songs_Spotify.csv
# Review the DataFrame
holiday_songs_spotify_df.head()
```

```python
#drop unneeded columns
holiday_songs_features_df = holiday_songs_spotify_df.drop(['Unnamed: 0','key_mode','time_signature', 'playlist_name', 'playlist_img',
                                                           'track_name', 'artist_name','album_name', 'album_img', 'track_uri', 
                                                          'key', 'loudness', 'mode','speechiness','acousticness', 'instrumentalness',
                                                           'tempo', 'duration_ms', 'time_signature','key_mode','liveness', 'valence'],axis=1)
holiday_songs_features_df.head()
```

```python
 # Create a a list to store inertia values and the values of k
inertia = []
k = list(range(2, 11))
```

```python
# Create a for-loop where each value of k is evaluated using the K-means algorithm
# Fit the model using the holiday_songs_spotify DataFrame
# Append the value of the computed inertia from the `inertia_` attribute of the KMeans model instance
for i in k:
    k_model = KMeans(n_clusters=i, random_state=1)
    k_model.fit(holiday_songs_features_df)
    inertia.append(k_model.inertia_)
```

```python
# Define a DataFrame to hold the values for k and the corresponding inertia
elbow_data = {"k": k, "inertia": inertia}
df_elbow = pd.DataFrame(elbow_data)

# Review the DataFrame
df_elbow.head()
```

```python
 # Plot the DataFrame
df_elbow.hvplot.line(
    x="k", 
    y="inertia", 
    title="Elbow Curve", 
    xticks=k
)
```

```python
#Using k-means with 3 clusters
```

```python
 # Define the model with 3 clusters
model = KMeans(n_clusters=3, random_state=1)

# Fit the model
model.fit(holiday_songs_features_df)

# Make predictions
k_3 = model.predict(holiday_songs_features_df)

# Create a copy of the DataFrame
holiday_songs_features_predictions_df = holiday_songs_features_df.copy()

# Add a class column with the labels
holiday_songs_features_predictions_df['features_segment'] =k_3
```

```python
 # Plot the clusters
holiday_songs_features_predictions_df.hvplot.scatter(
    x="danceability",
    y="energy",
    by="features_segment"
)
```

```python
 # Define the model with 5 clusters
model = KMeans(n_clusters=5, random_state=1)

# Fit the model
model.fit(holiday_songs_features_df)

# Make predictions
k_5 = model.predict(holiday_songs_features_df)

# Create a copy of the DataFrame
holiday_songs_features_predictions_df = holiday_songs_features_df.copy()

# Add a class column with the labels
holiday_songs_features_predictions_df['features_segment'] =k_5
```

```python
 # Plot the clusters
holiday_songs_features_predictions_df.hvplot.scatter(
    x="danceability",
    y="energy",
    by="features_segment"
)
```
